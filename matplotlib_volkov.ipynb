# 3.1 Несколько графиков на одной фигуре — plt.subplot()

```python
# Построй 2 графика в строку:

# слева — y = x²
# справа — y = x³
# Используй plt.subplot(1, 2, ...)
# Массив x = [1, 2, 3, 4]

x = np.array([1, 2, 3, 4])
plt.figure(figsize=(8, 5))
plt.subplot(1, 2, 1)
plt.plot(x, x**2)
plt.title('y = x²')

plt.subplot(1, 2, 2)
plt.plot(x, x**3)
plt.title('y = x³')

plt.tight_layout()
plt.show();
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
ax[0].plot(x, x**2, 'r')
ax[1].plot(x, x**3, 'b')
ax[0].set_title('y = x²')
ax[1].set_title('y = x³')
plt.tight_layout()
plt.show();
```

```python
# Создай 4 подграфика:
# y = x
# y = x²
# y = x³
# y = √x

# Сетка: 2 строки, 2 столбца
# Добавь заголовки, сделай выравнивание через plt.tight_layout().

x = np.array([1, 2, 3, 4])

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))
ax[0, 0].plot(x, x, 'r')
ax[0, 1].plot(x, x**2, 'g')
ax[1, 0].plot(x, x**3, 'b')
ax[1, 1].plot(x, x**0.5, 'y')

ax[0, 0].set_title('y = x')
ax[0, 1].set_title('y = x²')
ax[1, 0].set_title('y = x³')
ax[1, 1].set_title('y = √x')

plt.tight_layout()
plt.show();
```

```python
# Построй 3 подграфика в строку:
# y = x
# y = 2x
# y = 3x

# Добавь к каждому:
# Цвет линии
# Заголовок
# Подписи осей

x = np.array([1, 2, 3, 4])

fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(10, 5))
ax[0].plot(x, x, 'r')
ax[1].plot(x, 2*x, 'g')
ax[2].plot(x, 3*x, 'b')

ax[0].set_title('y = x')
ax[1].set_title('y = 2x')
ax[2].set_title('y = 3x')

for i in range(3):
    ax[i].set_xlabel('x')
    ax[i].set_ylabel('y', rotation=False)

plt.tight_layout()
plt.show();
```

```python
# Сделай сетку 2×3, и построй в каждой ячейке график y = x * n, где n от 1 до 6.
# Добавь заголовок: y = x * n, соответствующий коэффициенту.
# В конце вызови plt.tight_layout() и plt.show().

x = np.array([1, 2, 3, 4])

fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(10, 5))

n = 1
d = {k: v for k, v in zip(range(1, 7), 'rgbyrg')}

for i in range(2):
    for j in range(3):
        ax[i, j].plot(x, x * n, color=d.get(n))
        ax[i, j].set_title(f'y = x * {n}')
        n += 1

plt.tight_layout()
plt.show();
```

# 4.1 Диаграмма рассеяния — plt.scatter()

```python
x = np.random.rand(100)
y = np.random.rand(100)
sizes = np.random.rand(100) * 100
colors = np.random.rand(100)

plt.scatter(x, y, s=sizes, c=colors, alpha=0.6, cmap='viridis')
plt.colorbar(label="Цветовое значение")
plt.title("Продвинутая диаграмма рассеяния")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.tight_layout()
plt.show();
```

# 4.2 Столбчатая диаграмма — plt.bar()

```python
labels = ['A', 'B', 'C', 'D']
values = [10, 24, 36, 18]

colors = ['#FF5733', '#33FF57', '#3357FF', '#FFC300']

plt.bar(labels, values, color=colors, width=0.5)
plt.title("Цветная столбчатая диаграмма")
plt.xlabel("Категории")
plt.ylabel("Значения")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show();
```

```python
plt.barh(labels, values, color='skyblue')
plt.title("Горизонтальная диаграмма")
plt.xlabel("Значения")
plt.ylabel("Категории")
plt.grid(axis='x')
plt.tight_layout()
plt.show();
```

# 4.3 Гистограмма — plt.hist()

```python
data = np.random.randn(1000)

plt.hist(data, bins=20, color='skyblue', edgecolor='black')
plt.title("Гистограмма распределения")
plt.xlabel("Значения")
plt.ylabel("Частота")
plt.grid(True)
plt.show();
```

```python
data = np.random.normal(loc=5, scale=2, size=500)

plt.hist(data, bins=30, density=True, alpha=0.7, color='orange', edgecolor='black')
plt.title("Гистограмма плотности")
plt.xlabel("Значения")
plt.ylabel("Плотность")
plt.grid(True)
plt.show();
```

# 4.4 Диаграмма размаха (boxplot) — plt.boxplot()

```python
np.random.seed(0)

group1 = np.random.normal(5, 1, 500)
group2 = np.random.normal(6, 1.5, 500)
group3 = np.random.normal(7, 1, 500)

data = [group1, group2, group3]

plt.boxplot(data, patch_artist=True)
plt.title("Сравнение трёх выборок")
plt.xticks([1, 2, 3], ['Группа 1', 'Группа 2', 'Группа 3'])
plt.ylabel("Значения")
plt.grid(True)
plt.show();
```

```python
# Построй диаграмму размаха для трёх групп
# Добавь отображение среднего значения (showmeans=True) и закрась график.

group1 = np.random.normal(4, 1, 200)
group2 = np.random.normal(5, 1, 200)
group3 = np.random.normal(6, 1, 200)

data = [group1, group2, group3]

plt.boxplot(data, showmeans=True, patch_artist=True)
plt.title("Сравнение трёх выборок")
plt.xticks([1, 2, 3], ['Группа 1', 'Группа 2', 'Группа 3'])
plt.axhline(group2.mean(), color='C4')

plt.show();
```
