# 3.1 Несколько графиков на одной фигуре — plt.subplot()

```python
# Построй 2 графика в строку:

# слева — y = x²
# справа — y = x³
# Используй plt.subplot(1, 2, ...)
# Массив x = [1, 2, 3, 4]

x = np.array([1, 2, 3, 4])
plt.figure(figsize=(8, 5))
plt.subplot(1, 2, 1)
plt.plot(x, x**2)
plt.title('y = x²')

plt.subplot(1, 2, 2)
plt.plot(x, x**3)
plt.title('y = x³')

plt.tight_layout()
plt.show();
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
ax[0].plot(x, x**2, 'r')
ax[1].plot(x, x**3, 'b')
ax[0].set_title('y = x²')
ax[1].set_title('y = x³')
plt.tight_layout()
plt.show();
```

```python
# Создай 4 подграфика:
# y = x
# y = x²
# y = x³
# y = √x

# Сетка: 2 строки, 2 столбца
# Добавь заголовки, сделай выравнивание через plt.tight_layout().

x = np.array([1, 2, 3, 4])

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))
ax[0, 0].plot(x, x, 'r')
ax[0, 1].plot(x, x**2, 'g')
ax[1, 0].plot(x, x**3, 'b')
ax[1, 1].plot(x, x**0.5, 'y')

ax[0, 0].set_title('y = x')
ax[0, 1].set_title('y = x²')
ax[1, 0].set_title('y = x³')
ax[1, 1].set_title('y = √x')

plt.tight_layout()
plt.show();
```

```python
# Построй 3 подграфика в строку:
# y = x
# y = 2x
# y = 3x

# Добавь к каждому:
# Цвет линии
# Заголовок
# Подписи осей

x = np.array([1, 2, 3, 4])

fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(10, 5))
ax[0].plot(x, x, 'r')
ax[1].plot(x, 2*x, 'g')
ax[2].plot(x, 3*x, 'b')

ax[0].set_title('y = x')
ax[1].set_title('y = 2x')
ax[2].set_title('y = 3x')

for i in range(3):
    ax[i].set_xlabel('x')
    ax[i].set_ylabel('y', rotation=False)

plt.tight_layout()
plt.show();
```

```python
# Сделай сетку 2×3, и построй в каждой ячейке график y = x * n, где n от 1 до 6.
# Добавь заголовок: y = x * n, соответствующий коэффициенту.
# В конце вызови plt.tight_layout() и plt.show().

x = np.array([1, 2, 3, 4])

fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(10, 5))

n = 1
d = {k: v for k, v in zip(range(1, 7), 'rgbyrg')}

for i in range(2):
    for j in range(3):
        ax[i, j].plot(x, x * n, color=d.get(n))
        ax[i, j].set_title(f'y = x * {n}')
        n += 1

plt.tight_layout()
plt.show();
```

# 4.4 Диаграмма размаха (boxplot) — plt.boxplot()

```python
np.random.seed(0)

group1 = np.random.normal(5, 1, 500)
group2 = np.random.normal(6, 1.5, 500)
group3 = np.random.normal(7, 1, 500)

data = [group1, group2, group3]

plt.boxplot(data, patch_artist=True)
plt.title("Сравнение трёх выборок")
plt.xticks([1, 2, 3], ['Группа 1', 'Группа 2', 'Группа 3'])
plt.ylabel("Значения")
plt.grid(True)
plt.show();
```

```python
# Построй диаграмму размаха для трёх групп
# Добавь отображение среднего значения (showmeans=True) и закрась график.

group1 = np.random.normal(4, 1, 200)
group2 = np.random.normal(5, 1, 200)
group3 = np.random.normal(6, 1, 200)

data = [group1, group2, group3]

plt.boxplot(data, showmeans=True, patch_artist=True)
plt.title("Сравнение трёх выборок")
plt.xticks([1, 2, 3], ['Группа 1', 'Группа 2', 'Группа 3'])
plt.axhline(group2.mean(), color='C4')

plt.show();
```

# 4.6 Стековый график — plt.stackplot()

```python
years = [2015, 2016, 2017, 2018, 2019]
frontend = [10, 15, 20, 25, 30]
backend = [5, 10, 15, 20, 25]
devops = [2, 5, 8, 12, 15]

plt.stackplot(years, frontend, backend, devops,
              labels=["Frontend", "Backend", "DevOps"],
              colors=["#1f77b4", "#ff7f0e", "#2ca02c"],
              alpha=0.8)

plt.legend(loc='upper left')
plt.title("Рост команд по направлениям")
plt.xlabel("Год")
plt.ylabel("Количество разработчиков")
plt.grid(True)
plt.show();
```

```python
hours = list(range(6, 24))

mobile = [10, 20, 35, 50, 60, 65, 70, 80, 60, 50, 40, 30, 20, 10, 5, 2, 1, 0]
desktop = [5, 10, 15, 20, 30, 35, 40, 45, 40, 35, 30, 20, 10, 5, 3, 1, 0, 0]
tablet = [2, 4, 6, 8, 10, 12, 15, 20, 18, 15, 12, 10, 8, 6, 4, 2, 1, 0]

colors = ["#ff9999", "#66b3ff", "#99ff99"]
legend = ["Мобильные", "Десктоп", "Планшеты"]

plt.stackplot(hours, mobile, desktop, tablet, alpha=0.8)
plt.legend(legend)
plt.title("Доход от продуктов по годам")
plt.xlabel('hours')
plt.ylabel('count')
plt.grid()

plt.show();
```

# 4.7 Stem-диаграмма — plt.stem()

```python
x = [1, 2, 3, 4, 5]
y = [10, 7, 12, 8, 9]

plt.stem(x, y,
         linefmt='g-',        # зелёные линии
         markerfmt='ro',      # красные кружки
         basefmt='k-')        # чёрная базовая линия
plt.xlabel("Номер точки")
plt.ylabel("Значение")
plt.title("Stem-диаграмма простого ряда")
plt.grid(True)

plt.show();
```

```python
# Построй stem-диаграмму для следующих данных:

x = [0, 1, 2, 3, 4, 5]
y = [5, 3, 4, 2, 6, 1]
            
# Стилизация:
# Линии: зелёные ('g-')
# Маркеры: фиолетовые кружки ('mo')
# Базовая линия: чёрная ('k-')
# Заголовок: 'График сигнала'
# Подписи осей: Ось X, Значения
# Сетку

plt.stem(x, y, linefmt='g', markerfmt='mo', basefmt='k-')
plt.title('График сигнала')
plt.xlabel('Ось X')
plt.ylabel('Значения')
plt.grid()

plt.show();
```

```python
# У тебя есть предсказания и истинные значения. Построй stem-график ошибок модели (остатков).

y_true = np.array([3, 5, 2, 7, 4])
y_pred = np.array([2, 5, 3, 6, 5])
X = [0, 1, 2, 3, 4]

# Вычисли остатки: ошибка = y_true - y_pred
# Построй stem-график ошибок по оси X = [0, 1, 2, 3, 4]
# Оформление по умолчанию

loss = y_true - y_pred

plt.stem(X, loss)
plt.grid()

plt.show();
```

```python
# У тебя есть следующие значения частот и их амплитуд после преобразования Фурье:

frequencies = [1, 2, 3, 4, 5, 6]
amplitudes = [0.2, 1.1, 0.9, 0.4, 1.3, 0.7]
        
# Построй stem-диаграмму:
# Цвет линий: синий ('b-')
# Маркеры: красные ('ro')
# Базовая линия: серый ('--')
# Добавь заголовок "Амплитуды частот", подпиши оси: "Частота" и "Амплитуда".

plt.stem(frequencies, amplitudes, linefmt='b-', markerfmt='ro', basefmt='--')
plt.title('Амплитуды частот')
plt.xlabel('Частота')
plt.ylabel('Амплитуда')
plt.grid()

plt.show();
```

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 1, 3, 5]
y2 = [1, 3, 2, 5, 4]

plt.stem(x, y1, linefmt='g-', markerfmt='bo')
plt.grid()

plt.show();
```

# 4.8 Ступенчатый график — plt.step()

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 0, 1, 0, 1]

plt.step(x, y)
plt.title("Ступенчатый график (по умолчанию: where='pre')")
plt.grid(True)

plt.show();
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 2, 1]

plt.step(x, y, where='pre', label='pre')
plt.step(x, y, where='post', label='post')
plt.step(x, y, where='mid', label='mid')

plt.legend()
plt.title("Сравнение различных режимов 'where'")
plt.grid(True)

plt.show();
```

# 4.9 График с погрешностями — plt.errorbar()

```python
x = [1, 2, 3, 4]
y = [2.5, 3.6, 3.0, 4.5]
errors = [0.2, 0.3, 0.15, 0.25]

plt.errorbar(x, y, yerr=errors, fmt='o', ecolor='red', capsize=5)
plt.title("Результаты эксперимента с погрешностями")
plt.xlabel("Время (сек)")
plt.ylabel("Значение")
plt.grid(True)

plt.show();
```

```python
lower = [0.2, 0.1, 0.2, 0.15]
upper = [0.4, 0.3, 0.25, 0.35]

plt.errorbar(x, y, yerr=[lower, upper], fmt='o', capsize=4, ecolor='blue')
plt.title("Асимметричные погрешности")
plt.grid(True)

plt.show();
```

```python
xerr = [0.1, 0.2, 0.1, 0.3]

plt.errorbar(x, y, xerr=xerr, yerr=errors,
             fmt='s', capsize=4, ecolor='green')
plt.title("Погрешности по обеим осям")
plt.grid(True)

plt.show();
```

```python
x = [1, 2, 3, 4]
y = [1.5, 2.5, 2.0, 3.5]
xerr = [0.1, 0.2, 0.1, 0.3]
yerr = [0.2, 0.1, 0.3, 0.2]

plt.errorbar(x, y, xerr=xerr, yerr=errors, fmt='gs', capsize=4, ecolor='black')

plt.show();
```

# 4.10 Контурный график — plt.contour()

```python
# Создаем сетку координат
x = np.linspace(-6, 6, 100)
y = np.linspace(-6, 6, 100)
X, Y = np.meshgrid(x, y)

# Вычисляем значения функции
Z = np.sin(np.sqrt(X**2 + Y**2))

# Строим контурный график
plt.contour(X, Y, Z, levels=15, colors='blue')

plt.title('Контурный график функции sin(sqrt(x² + y²))')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)

plt.show();
```

```python
x = np.array([-3, 3, 0.1])
y = np.array([-3, 3, 0.1])
X, Y = np.meshgrid(x, y)
Z = np.sin(X**2 + Y**2)

plt.contourf(X, Y, Z, levels=30)
plt.grid()

plt.show();
```
