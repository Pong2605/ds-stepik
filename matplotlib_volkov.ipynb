# 2.1 Построение простого графика

```python
x = [1, 2, 3]
y = [3, 2, 1]

plt.plot(x, y, color='black', label='Данные A',)
plt.grid(color='grey', linestyle=':')
plt.legend(loc='lower left')
plt.show();
```

```python
plt.plot(x, y, label='Данные', color='blue')
plt.legend(loc='upper right', fontsize=12, shadow=True)
plt.title('График с легендой')
plt.show()
```
