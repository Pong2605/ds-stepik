# 1.2 scatterplot

```python
# Основные параметры этой функции:

# data: DataFrame или массив с данными;
# x: название столбца или вектор данных для оси X;
# y: название столбца или вектор данных для оси Y;
# hue: название столбца или вектор данных для группировки точек по цвету;
# size: название столбца или вектор данных для изменения размера точек;
# style: название столбца или вектор данных для изменения стиля точек (например, разные формы);
# palette: цветовая палитра для переменной hue;
# hue_order: задаёт порядок отображения категорий для переменной hue;
# hue_norm: позволяет нормализовать данные переменной hue;
# sizes: диапазон размеров для переменной size;
# size_order: задаёт порядок отображения категорий для переменной size;
# size_norm: позволяет нормализовать данные переменной size, задавая диапазон значений;
# markers: список стилей маркеров для переменной style;
# style_order: задаёт порядок отображения категорий для переменной style.
```

# 1.3 lineplot - линейный график

```python
exercise = sns.load_dataset('exercise')
exercise = exercise.drop('Unnamed: 0', axis=1)

exercise

# 'id' - уникальный идентификатор участника тестирования;
# 'diet' - диета, которую придерживался участник (no fat - полное исключение жиров, low fat - низкожировая диета);
# 'pulse' - средний пульс участника после выполнения упражнения;
# 'time' - время, проведенное на упражнениях (в минутах);
# 'kind' - тип упражнения (running (бег), walking (ходьба), rest (отдых)).
```

```python
exercise.info()
```

```python
sns.lineplot(x='time', y='pulse', data=exercise)

plt.grid(linestyle='--', alpha=0.5);
```

```python
# В итоге мы получили линейный график с некоторой областью, выделенной голубым цветом. Эта область называется доверительным интервалом. 
# Отключим эту опцию, передав функции sns.lineplot параметр ci (или errorbar) со значением None (ci=None)

sns.lineplot(x='time', y='pulse', data=exercise,
             errorbar=None,             # - отключить отображение доверительного интервала
             marker='s',                # - тип маркера
             markersize=10,             # - размер маркера
             markerfacecolor='red',     # - цвет меток
             linewidth=4,               # - ширина линий
             color='green',             # - цвет линий
             alpha=0.6)                 # - прозрачность линий и меток

plt.grid(linestyle='--', alpha=0.5);
```

```python
# Параметры hue, hue_order

sns.lineplot(x='time', y='pulse', data=exercise,
             hue='kind',                                    # - разделение признака/значения на несколько линий
             hue_order=['rest', 'running', 'walking'])      # - список и порядок значений параметров

sns.set_style('white');
```

```python
# Параметры size, size_order и sizes

sns.lineplot(x='time', y='pulse', data=exercise,
             hue='kind',
             errorbar=None,
             alpha=0.5,
             size='kind',                                   # - указать переменную, значения которой будут определять толщину линий на графике
             size_order=['running', 'walking', 'rest'],     # - задать конкретный порядок для значений переменной, указанной в size
             sizes=(2, 4))                                  # - задать диапазон возможных размеров линий

plt.grid(linestyle='--', alpha=0.5);
```

```python
# Параметр estimator - выбор метода агрегации данных в функции sns.lineplot()

sns.lineplot(x="time", y="pulse", data=exercise,
             hue='kind',
             errorbar=None,
             marker='o',
             estimator='median')

plt.grid(linestyle='--', alpha=0.5);
```

```python
sns.lineplot(x="time", y="pulse", data=exercise,
             hue='kind',
             errorbar=None,
             marker='o',
             estimator=np.sum)

plt.grid(linestyle='--', alpha=0.5);
```

```python
# Параметр ci (errorbar)

# Доверительный интервал (Confidence Interval, CI) - это диапазон значений, который, как предполагается, 
# содержит истинное значение параметра с определённым уровнем достоверности. 
# Например, 95%-й доверительный интервал означает, что если бы мы многократно повторяли эксперименты и строили доверительные интервалы, 
# то в 95% случаев этот интервал содержал бы истинное значение параметра.
```

```python
# По умолчанию ci=95, что означает построение 95%-го доверительного интервала вокруг средней линии.

sns.lineplot(x='time', y='pulse', data=exercise,
             hue='diet')

plt.grid(linestyle='--', alpha=0.5);
```

```python
# Сузим доверительный интервал до 40%

sns.lineplot(x='time', y='pulse', data=exercise,
             hue='diet',
             errorbar=('ci', 40))

plt.grid(linestyle='--', alpha=0.5);
```

```python
# Параметр errorbar позволяет отображать на графике не только доверительный интервал, но и другие статистически полезные данные, 
# такие как различные персентили, стандартную ошибку и кастомные функции (min, max).

sns.lineplot(x='time', y='pulse', data=exercise,
             hue='diet',
             errorbar=('se'),
             palette='hot')

plt.grid(linestyle='--', alpha=0.5);
```

# 1.4 relplot - универсальной функцией для визуализации отношений между переменными

```python
# Используется для создания различных типов реляционных графиков, таких как линейные графики (lineplot) и точечные графики (scatterplot).
```

```python
# Основные параметры функции :

# data: DataFrame, массив с данными;
# x, y: имена переменных, которые будут отображены на осях X и Y, соответственно;
# hue: переменная для группировки данных по цвету;
# size: переменная, значение которой будет использоваться для изменения размера точек;
# style: переменная, значение которой будет использоваться для изменения стиля точек или линий;
# units: имя переменной, которая определяет семантические подгруппы для раздельного отображения каждой точки;
# weights: имя переменной, которая определяет веса точек (используется при построении регрессии или других статистических моделей);
# row: имя переменной, которая определяет, как данные будут разбиты на строки при использовании нескольких подграфиков;
# col: имя переменной, которая определяет, как данные будут разбиты на столбцы при использовании нескольких подграфиков;
# col_wrap: количество столбцов в сетке подграфиков до переноса на новую строку;
# row_order: порядок, в котором строки должны быть отображены;
# col_order: порядок, в котором столбцы должны быть отображены;
# palette: параметр, определяющий цветовую палитру для переменной hue. Может быть именем палитры, списком цветов или словарем, отображающим уровни переменной в цвета;
# hue_order: порядок отображения уровней переменной hue;
# hue_norm: нормализация для цветовой палитры;
# sizes: значения для переменной size, которые будут отображены в виде размеров маркеров. Может быть словарем или списком значений;
# size_order: порядок отображения уровней переменной size;
# size_norm: нормализация для размеров маркеров;
# markers: маркеры для переменной style. Может быть списком или словарем;
# dashes: определяет типы линий (например, пунктирные) для переменной style;
# style_order: определяет порядок отображения уровней переменной style;
# legend: управляет отображением легенды ('auto', 'brief', 'full', или False);
# kind: тип графика. Может быть 'scatter' (по умолчанию) для диаграммы рассеяния или 'line' для линейного графика;
# height: управляет высотой каждого подграфика (по умолчанию height=5);
# aspect: соотношение ширины и высоты каждого подграфика (по умолчанию aspect=1).
```

```python
taxis = sns.load_dataset('taxis').dropna()

taxis.head()

# 'pickup': Дата и время, когда пассажир был подобран.
# 'dropoff': Дата и время, когда пассажир был высажен.
# 'passengers': Количество пассажиров в поездке.
# 'distance': Расстояние поездки, измеренное в милях.
# 'fare': Стоимость проезда без учета чаевых и сборов, измеренная в долларах.
# 'tip': Чаевые, данные водителю, измеренные в долларах.
# 'tolls': Плата за проезд по платным дорогам, измеренная в долларах.
# 'total': Общая стоимость поездки, включающая проезд, чаевые и сборы, измеренная в долларах.
# 'color': Цвет такси (желтое или зеленое).
# 'payment': Способ оплаты (наличные, кредитная карта).
# 'pickup_zone': Зона, в которой был подобран пассажир.
# 'dropoff_zone': Зона, в которой пассажир был высажен.
# 'pickup_borough': Район города, в котором был подобран пассажир (Manhattan, Queens, Bronx, Brooklyn).
# 'dropoff_borough': Район города, в котором пассажир был высажен.
```

```python
taxis.info()
```

```python
df = taxis[taxis['pickup_borough'] == 'Queens']

df.head()
```

```python
sns.relplot(data=df, x='pickup', y='total',
            aspect=2.2,                         # - расширить ось X
            hue='color',                        # - разделение по цвету такси
            palette=['yellow', 'green'])        # - выбор используемых цветов
plt.show()
```

```python
sns.relplot(data=df, x='pickup', y='total')
plt.xticks(rotation=90)                     # - поворот меток
plt.show()
```

```python
sns.relplot(data=df, x=taxis['pickup'].dt.day_name(), y='total',
            aspect=1.3,
            hue='color',
            palette=['yellow', 'green'])
plt.show()
```

```python
# Для того, чтобы отобразить дни недели на графике по порядку, проведём следующие манипуляции. 
# Сначала создадим дополнительный столбец 'pickup_day_name', содержащий дни недели.
# Далее сделаем этот столбец категориальным с определённым порядком следования категорий.
# Параметр categories задаёт фиксированный список категорий, а ordered=True указывает, что категории упорядочены.

df['pickup_day_name'] = df['pickup'].dt.day_name()
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['pickup_day_name'] = pd.Categorical(df['pickup_day_name'], categories=day_order, ordered=True)
```

```python
sns.relplot(data=df, x='pickup_day_name', y='total',
            aspect=1.3,
            hue='color',
            palette=['yellow', 'green'])
plt.show()
```

```python
# Параметр kind='line' позволяет изменить тип графика на линейный

sns.relplot(data=df, x='pickup_day_name', y='total',
            aspect=1.3,
            hue='color',
            palette=['yellow', 'green'],
            kind='line')
plt.show()
```

```python
sns.relplot(data=df, x='pickup_day_name', y='total',
            aspect=1.3,
            hue='color',
            palette=['yellow', 'green'],
            kind='line',
            errorbar=None,
            marker='s',
            style='color')
plt.show()
```

```python
sns.relplot(data=df, x='pickup_day_name', y='total',
            aspect=1.3,
            hue='color',
            palette=['yellow', 'green'],
            kind='line',
            style='color',
            dashes=[(2, 8), (6, 2)],        # - ширина штриха линий и пробела между ними
            linewidth=4)    
plt.show()
```

```python
sns.relplot(data=df, x='pickup', y='total',
            aspect=1.3,
            hue='color',
            palette=['yellow', 'green'],
            style='color',        # - управляет формой точек
            size='color')         # - определяет размер точек на графике в зависимости от значений определенной переменной
plt.xticks(rotation=45)
plt.show()
```

```python
g = sns.relplot(data=taxis, x='pickup', y='total',
                height=4,
                hue='color',
                palette=['yellow', 'green'],
                col='pickup_borough')

# Результат выполнения функции sns.relplot() сохраняется в переменной g, что позволяет впоследствии применять к ней различные методы
# Далее используется цикл с вложенным циклом: g.axes.flat - это итератор, позволяющий перебрать все подграфики в последовательности. 
# То есть сначала переменная ax принимает результат выполнения первого подграфика. Затем происходит выполнение вложенного цикла:
# Метод get_xticklabels() позволяет получить доступ к меткам на оси X, чтобы их можно было настроить. 
# То есть переменная label последовательно принимает значения меток по оси X первого подграфика и методом set_rotation(45) поворачивает каждую метку на 45 градусов. 
# Когда все метки первого подграфика пройдены, переменная ax принимает результат выполнения второго подграфика и т.д.
for ax in g.axes.flat: 
    for label in ax.get_xticklabels(): 
        label.set_rotation(45)

plt.show()

```

```python
# Когда переменная, передаваемая параметру col, имеет много значений, удобно использовать параметр col_wrap, который управляет количеством столбцов в сетке графиков.

g = sns.relplot(data=taxis, x='pickup', y='total',
                height=4,
                hue='color',
                palette=['yellow', 'green'],
                col='pickup_borough',
                col_wrap=3)

for ax in g.axes.flat: 
    for label in ax.get_xticklabels(): 
        label.set_rotation(45)

plt.show()
```
