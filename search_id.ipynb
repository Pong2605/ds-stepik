```python
import pandas as pd
import numpy as np
```

```python
events_data = pd.read_csv('data\event_data_train.zip', compression='zip')
submissions_data = pd.read_csv('data\submissions_data_train.zip', compression='zip')
```

```python
events_data_sortid = events_data.sort_values(['user_id', 'date'])
```

```python
events_data_sortid['diff'] = events_data_sortid.groupby('user_id')['date'].diff(1)
```

```python
events_data_sortid
```

```python
events_data_start = events_data_sortid[(events_data_sortid['action'] == 'discovered')].drop_duplicates(subset=['step_id', 'user_id'])
```

```python
events_data_stop = events_data_sortid[(events_data_sortid['action'] == 'passed')].drop_duplicates(subset=['step_id', 'user_id'])
```

```python
events_data_start
```

```python
events_data_stop
```

```python
events_data_full = pd.concat([events_data_start, events_data_stop]).sort_values(['user_id', 'date'])
```

```python
events_data[events_data['action'] == 'passed'].groupby('user_id').agg({'step_id' : 'count'}).reset_index().rename(columns={'step_id' : 'count'})
```

```python
events_data_passed = events_data[events_data['action'] == 'passed']
```

```python
events_data.groupby('action').count()
```

```python
max_count = events_data[events_data['action'] == ('passed')].groupby('user_id').size().max()
```

```python
tmp_df = events_data[events_data['action'] == 'passed'].groupby('user_id').filter(lambda x: x['step_id'].nunique() == max_count)
```

```python
tmp_df.user_id.nunique()
```

```python
len(tmp_df) / len(events_data_passed)
```

```python
events_data_top_users = events_data[events_data['user_id'].isin(tmp_df['user_id'])]
```

```python
events_data_top_users
```

```python
top_users_start = events_data_top_users[events_data_top_users.action == 'discovered'].sort_values('user_id')
```

```python
top_users_start.head()
```

```python
top_users_stop = events_data_top_users[events_data_top_users.action == 'passed'].sort_values('user_id')
```

```python
top_users_stop.head()
```

```python
top_users_full = pd.concat([top_users_start, top_users_stop]).sort_values(['user_id', 'date'])
```

```python
top_users_full['diff'] = top_users_full.groupby('user_id')['timestamp'].diff(1).fillna(0)
```
