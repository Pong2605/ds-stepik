# Подключение к базе

```python
con = sqlite3.connect('db')
```

```python
df = pd.DataFrame({'col_1': [1, 2, 3]})
```

```python
df
```

```python
df.to_sql(name='test_table', con=con, index=False, if_exists='replace')
```

```python
sql = '''SELECT * FROM test_table'''
```

```python
pd.read_sql(sql=sql, con=con)
```

```python
pd.read_sql(sql=sql, con=con).plot();
```

```python
sql = '''SELECT 
t.col_1 * 3 AS multiply_col_1
FROM test_table t
LIMIT 2'''

pd.read_sql(sql, con)
```

```python
def select(sql):
    return pd.read_sql(sql, con)
```

```python
select(sql)
```
