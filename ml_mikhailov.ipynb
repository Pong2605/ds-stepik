## 3.1  Человеческое обучение

```python
df = pd.read_csv(r'data\uk_used_car_data_set\bmw.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(df, random_state=42, test_size=0.25)
```

```python
mean_median_price_model = train.groupby('year')['price'].agg(['mean', 'median']).reset_index().rename(columns={'mean': 'price_mean', 'median': 'price_median'})
```

```python
mean_median_price_model
```

```python
mean_median_price_model = train.merge(mean_median_price_model, how='left', on='year')
```

```python
mean_median_price_model
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error
```

```python
mean_absolute_error(mean_median_price_model['price'], mean_median_price_model['price_mean'])
```

```python
mean_absolute_percentage_error(mean_median_price_model['price'], mean_median_price_model['price_mean'])
```

```python
mean_absolute_error(mean_median_price_model['price'], mean_median_price_model['price_median'])
```

```python
mean_absolute_percentage_error(mean_median_price_model['price'], mean_median_price_model['price_median'])
```
