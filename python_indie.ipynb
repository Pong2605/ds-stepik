# Парсинг строки через literal_eval

```python
import ast

input_str = '''1, 2.5, "hello", [1, 2, 3], "world", 5'''

parsed_list = ast.literal_eval(f"[{input_str}]")
unique_types = {type(item) for item in parsed_list}

len(unique_types)
```
