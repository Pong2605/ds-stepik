# Рекурсия

```python
# Дано число и количество раз, которое его нужно умножить на себя. 
# Необходимо написать рекурсивную функцию multiply(num, times), 
# которая умножает число само на себя заданное целое натуральное количество раз (или 0 раз). 
# Каждый элемент списка multiply — это результат одного умножения. 
# Использовать встроенный оператор ** для возведения в степень запрещено. 
# Пользуйтесь только умножением *.

# Например, для числа 2 и количества умножений 4 результат вычисляется так:
# 2 умножить 4 раза = (((2 * 2) * 2) * 2) = 16

num, times = 3, 3

def multiply(num, times):
    if times == 0:
        return 1
    return num * multiply(num, times - 1)

res = multiply(num, times)
print(res)
```

```python
# Необходимо написать рекурсивную функцию factorial(k), которая вычисляет факториал числа k. 
# Каждый элемент списка factorial — это результат одного умножения на следующее число. 
# Использовать встроенные функции для вычисления факториала запрещено. 
# Пользуйтесь только умножением * и рекурсией.

# Например, для числа 5 факториал вычисляется так:
# 5! = 5 * 4 * 3 * 2 * 1 = 120

k = 5

def factorial(k):
    if k == 0 or k == 1:
        return 1
    return k * factorial(k - 1)

res = factorial(k)
print(res)
```

```python
# Дана строка, содержащая только английские буквы (большие и маленькие). 
# Необходимо написать функцию add_brackets(), которая принимает на вход строку. 

# Функция должна добавлять в строку открывающиеся и закрывающиеся скобки по следующему образцу:
# "example" → "e(x(a(m)p)l)e"

# До середины добавляются открывающие скобки, после середины — закрывающие. 
# Если длина строки равна чётному числу, в скобках, расположенных в середине, должно быть два символа.
```

```python
line = 'example'

def add_brackets(line):
    if len(line) <= 2:
        return line
    return line[0] + '(' + add_brackets(line[1: -1]) + ')' + line[-1]

add_brackets(line)
```

```python
# Чтение чисел из stdin
numbers = [45, 35, 25, 55, 15, 65, 95, 85, 105, 75]

# Допишите ваш код здесь

def selection_sort(numbers):
    n = len(numbers)
    for i in range(n - 1):
        min_index = i
        for j in range(i + 1, n):
            if numbers[j] < numbers[min_index]:
                min_index = j

        min_value = numbers.pop(min_index)
        numbers.insert(i, min_value)
            
selection_sort(numbers)

print(numbers)
```

```python
# Чтение чисел из stdin
numbers = [45, 35, 25, 55, 15, 65, 95, 85, 105, 75]

# Допишите ваш код здесь

def selection_sort(numbers):
    n = len(numbers)
    for i in range(n - 1):
        min_index = i
        for j in range(i + 1, n):
            if numbers[j] < numbers[min_index]:
                numbers[min_index], numbers[j] = numbers[j], numbers[min_index]
            
selection_sort(numbers)

print(numbers)
```

```python
# Чтение чисел из stdin
numbers = [45, 35, 25, 55, 15, 65, 95, 85, 105, 75]

# Допишите ваш код здесь

def insertion_sort(numbers):
    n = len(numbers)
    for i in range(1, n):
        insert_index = i
        current_value = numbers[i]
        for j in range(i - 1, -1, -1):
            if numbers[j] > current_value:
                numbers[j + 1] = numbers[j]
                insert_index = j
            else:
                break
        numbers[insert_index] = current_value
            
insertion_sort(numbers)

print(numbers)
```
