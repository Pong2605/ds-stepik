```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
# Matplotlib

# Matplotlib - это основная библиотека для создания визуализаций в Python. Она предоставляет низкоуровневый интерфейс для создания графиков и имеет следующие характеристики:

# Гибкость и мощь: Matplotlib позволяет создавать сложные и детализированные графики. Она предоставляет полный контроль над всеми аспектами графика, включая стилизацию, размеры, цвета и расположение элементов.
# Сложность: Из-за своей гибкости, Matplotlib может быть сложной для освоения новичками, так как требует написания большого количества кода для создания и настройки графиков.
# Широкий спектр графиков: Matplotlib поддерживает создание различных типов графиков, таких как линейные графики, гистограммы, столбчатые графики, диаграммы рассеяния, диаграммы размаха и многие другие.
# Словосочетание "низкоуровневый интерфейс", применительно к библиотеке Matplotlib, обозначает ее способность предоставлять детальный и полный контроль над элементами графиков. Это означает, что пользователи могут настроить практически каждый аспект графика, от базовых элементов (таких как линии и точки) до более сложных компонентов (таких как аннотации, сетки и легенды).

# Отметим также, что библиотеки Seaborn и Pandas, в контексте визуализации данных, относятся к библиотекам с "высокоуровневым интерфейсом", которые предоставляют пользователям упрощенные и интуитивно понятные функции для создания графиков с минимальным количеством кода и настроек. Эти библиотеки автоматизируют многие аспекты создания графиков, позволяя пользователям быстро и легко визуализировать данные.

# Когда использовать Matplotlib (с определённой долей субъективности, т.к. каждый сам для себя определяет каким инструментарием ему пользоваться при построении графиков и диаграмм):

# Когда требуется полный контроль над внешним видом графиков.
# Для создания графиков с высокой степенью кастомизации (настройки) и детализации.
# В случаях, когда нужны нестандартные типы графиков или специфические настройки.
```

# 2.2 Строим график

```python
x = np.linspace(-np.pi, np.pi, 20)
cos_x = np.cos(x) 
sin_x = np.sin(x) 

print(x, cos_x, sin_x, sep='\n\n')
```

```python
# Создадим объекты Figure и Axes и построим первый график:

fig = plt.figure(figsize=(7,5))
ax = fig.add_axes([0, 0, 1, 1])

ax1 = ax.plot(x, cos_x) 
ax2 = ax.plot(x, sin_x)

plt.show()
```

# 2.3 Легенда, линии сетки и цвет фона

```python
# Работа с функцией Axes.legend() аналогична работе с plt.legend()

fig = plt.figure(figsize=(7,5))
ax = fig.add_axes([0, 0, 1, 1])

ax1 = ax.plot(x, cos_x) 
ax2 = ax.plot(x, sin_x)

ax.legend(labels=['cos(x)', 'sin(x)'], loc='upper left', ncols=2,                   # ncols - количество колонок в легенде
          facecolor='lightgrey', edgecolor='red', labelcolor='red', fontsize=12)    # настройки фона и шрифтов легенды

plt.show()
```

```python
# Работа с функцией Axes.grid() аналогична работе с plt.grid()

fig = plt.figure(figsize=(7,5))
ax = fig.add_axes([0, 0, 1, 1])

ax1 = ax.plot(x, cos_x, marker='o', c='c', linewidth=2) 
ax2 = ax.plot(x, sin_x, marker='<', c='r', linewidth=2)

ax.legend(labels=['cos(x)', 'sin(x)']) 
ax.grid(color='w')
ax.set_facecolor(color='#FAF7E3') # устанавливаем цвет заливки

plt.show()
```

# 2.4 Заголовок, названия и размерности осей

```python
# При объектно-ориентированном подходе к построению графиков мы будем пользоваться функциям 'plt.***' добавляется приставка set_, например set_title().
```

```python
fig = plt.figure(figsize=(7,5))
ax = fig.add_axes([0, 0, 1, 1])

ax1 = ax.plot(x, cos_x, marker='o', c='c', linewidth=2) 
ax2 = ax.plot(x, sin_x, marker='*', c='r', linewidth=2)

ax.legend(labels=['cos(x)', 'sin(x)']) 
ax.grid(alpha=0.3)

ax.set_title('Графики тригонометрических\nфункций', fontsize=15, fontweight ='bold', color='c', pad=15, loc='right') # \n - перенос на следующую строку

plt.show()

# Параметр pad в функции Axes.set_title() управляет расстоянием (в пикселях) между заголовком графика и самой осью.
```

```python
# Функции Axes.set_xlabel() и Axes.set_ylabel() в библиотеке matplotlib используются для установки меток (подписей) осей графика: 
# горизонтальной (ось X) и вертикальной (ось Y), соответственно.
# Функции Axes.set_xlim() и Axes.set_ylim() в библиотеке matplotlib используются для для изменения размерности осей графика.
# Функциями Axes.set_xticks() и Axes.set_yticks() можно принудительно устанавливать значения на осях графика по своему усмотрению.


fig = plt.figure(figsize=(7,5))
ax = fig.add_axes([0, 0, 1, 1])

ax1 = ax.plot(x, cos_x, marker='o', c='c', linewidth=2) 
ax2 = ax.plot(x, sin_x, marker='*', c='r', linewidth=2)

ax.legend(labels=['cos(x)', 'sin(x)']) 
ax.grid(alpha=0.3)
ax.set_title('Графики тригонометрических функций', fontsize=15, fontweight ='bold', color='c')

ax.set_xlabel('х', fontsize=12, fontweight ='bold', color='c')
ax.set_ylabel('Функция от х', fontsize=12, fontweight ='bold', color='c')

ax.set_xlim(-4, 4)
ax.set_ylim(-1.5, 1.5)

ax.set_xticks([-4, -2, 0, 2, 4])

plt.show()
```

# 2.5 tick_params()

```python
# labelrotation - поворот меток
# labelcolor    - цвет меток
# color         - цвет рисок
# colors        - цвет рисок и меток
# labelsize     - размер меток
# pad           - расстояние между метками оси и границей самой координатной оси
# direction     - риски на координатных осях: 'out' (наружу), 'in' (внутрь), 'inout' (внутрь и наружу)
# length        - длина рисок
# width         - ширина рисок

fig = plt.figure(figsize=(7,5))
ax = fig.add_axes([0, 0, 1, 1])

ax1 = ax.plot(x, cos_x, marker='o', c='c', linewidth=2) 
ax2 = ax.plot(x, sin_x, marker='*', c='r', linewidth=2)

ax.tick_params(axis='x', labelrotation=45, labelcolor='red', direction='in')
ax.tick_params(axis='y', labelrotation=-45, pad=30, direction='inout')

ax.tick_params(labelsize=12)
ax.tick_params(color='purple', length=7, width=2)


plt.show()
```

```python
# Промежуточные отсечки на оси set_minor_locator()

from matplotlib.ticker import AutoMinorLocator
```

```python
fig = plt.figure(figsize=(7,5))
ax = fig.add_axes([0, 0, 1, 1])

ax1 = ax.plot(x, cos_x, marker='o', c='c', linewidth=2) 
ax2 = ax.plot(x, sin_x, marker='*', c='r', linewidth=2)

ax.xaxis.set_minor_locator(AutoMinorLocator(5))
ax.yaxis.set_minor_locator(AutoMinorLocator(10))

plt.show()
```

```python
# Параметр which функции Axes.tick_params() 'major', 'minor' и 'both'

fig = plt.figure(figsize=(7,5))
ax = fig.add_axes([0, 0, 1, 1])

ax1 = ax.plot(x, cos_x, marker='o', c='c', linewidth=2) 
ax2 = ax.plot(x, sin_x, marker='*', c='r', linewidth=2)

ax.xaxis.set_minor_locator(AutoMinorLocator(5))
ax.yaxis.set_minor_locator(AutoMinorLocator(5))

ax.tick_params(which='major', axis='x', pad=10)
ax.tick_params(which='minor', labelsize=7)
ax.xaxis.set_minor_formatter('{x:.1f}')
```

```python
fig = plt.figure(figsize=(7,5))
ax = fig.add_axes([0, 0, 1, 1])

ax1 = ax.plot(x, cos_x, marker='o', c='c', linewidth=2) 
ax2 = ax.plot(x, sin_x, marker='*', c='r', linewidth=2)

ax.xaxis.set_minor_locator(AutoMinorLocator(5))
ax.yaxis.set_minor_locator(AutoMinorLocator(5))

ax.tick_params(which='major', grid_linestyle=':', grid_linewidth=2)
ax.tick_params(which='major', axis='x', pad=10)
ax.tick_params(which='minor', labelsize=7, grid_alpha=0.2)
ax.xaxis.set_minor_formatter('{x:.1f}')
ax.grid(which='both')

plt.show()
```

# 2.7 GridSpec

```python
# Класс GridSpec библиотеки matplotlib предоставляет возможность компоновать подграфики в различных конфигурациях. 
# Перед использованием, его нужно импортировать следующей командой:

from matplotlib.gridspec import GridSpec
```

```python
# Сначала создаём объект Figure.
# Затем создаём экземпляр класса GridSpec, которому передаём аргументы, содержащие количество рядов и количество столбцов. 
# Также "связываем" экземпляр класса GridSpec с объектом Figure через параметр figure.
# Создаём оси координат функцией fig.add_subplot() (или функцией plt.subplot()), и в качестве аргумента передаём этой функции местоположение, 
# где будут размещены данные оси координат.

fig = plt.figure(figsize=(10, 5))

gs = GridSpec(nrows=2, ncols=2, figure=fig)

ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,0])
ax4 = fig.add_subplot(gs[1,1])

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(10, 5))

gs = GridSpec(nrows=3, ncols=3, figure=fig)

ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[0,2])
ax4 = fig.add_subplot(gs[1:,0:2])

plt.tight_layout()
plt.show()
```

```python
# width_ratios и height_ratios
# Например, если мы хотим, чтобы высота координатных осей второго ряда была в два раза больше высоты координатных осей первого ряда (и у нас всего два ряда), 
# то тогда параметру height_ratios передаём список [1, 2]. 
# И вдобавок к этому, мы хотим, чтобы ширина координатных осей второго столбца была в три раза больше ширины координатных осей первого столбца 
# (допустим, что у нас всего два столбца), и в этом случае параметру width_ratios передаём список [1, 3].

fig = plt.figure(figsize=(10, 5))

widths = [1, 3]
heights = [1, 2]

gs = GridSpec(nrows=2, ncols=2, figure=fig, width_ratios=widths, height_ratios=heights)

ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,0])
ax4 = fig.add_subplot(gs[1,1])

plt.tight_layout()
plt.show()
```

```python
# Используя срезы вместе с параметрами width_ratios и height_ratios, мы можем реализовывать самые разнообразные и сложные компоновки подграфиков на одном рисунке.

fig = plt.figure(figsize=(10, 5))

widths = [1, 2, 1]
heights = [1, 2, 1]

gs = GridSpec(nrows=3, ncols=3, figure=fig, width_ratios=widths, height_ratios=heights)

ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,0])
ax4 = fig.add_subplot(gs[1,1])
ax5 = fig.add_subplot(gs[:,2])
ax6 = fig.add_subplot(gs[2,0:2])

plt.tight_layout()
plt.show()
```

```python
# Размер каждого графика формируется следующим образом:.

# Ширина i-ого графика:
# width_ratios[i] / sum(width_ratios)

# Высота i-ого графика:
# height_ratios[i] / sum(height_ratios)

# [0, 0]     [0, 1]     [0, 2]

# [1, 0]     [1, 1]     [1, 2]

# [2, 0]     [2, 1]     [2, 2]
```

```python
x = np.linspace(-np.pi, np.pi, 30)
y1 = np.sin(x)
y2 = 0.5 * np.sin(x)
y3 = 2 * np.sin(x)

fig = plt.figure(figsize=(10, 5))

widths = [2, 2, 1]
heights = [1, 2, 2]

gs = GridSpec(nrows=3, ncols=3, figure=fig, width_ratios=widths, height_ratios=heights)

ax1 = fig.add_subplot(gs[0,0])
ax1.plot(x, y1, 'y')
ax2 = fig.add_subplot(gs[0,1])
ax2.plot(x, y2, 'g')
ax3 = fig.add_subplot(gs[1,:])
ax3.plot(x, y3, 'm')

plt.tight_layout()
plt.show()
```

# 3.1 scatter - диаграмма рассеяния

```python
# Диаграмма рассеяния (точечная диаграмма) - это диаграмма, изображающая значения двух переменных в виде точек на плоскости.
# Функция plt.scatter() принимает только два аргумента за раз - координаты x и y для одной группы точек.
```

```python
from sklearn.datasets import load_iris

iris_dataset = load_iris()
df = pd.DataFrame(iris_dataset['data'], columns=iris_dataset['feature_names'])
df['target'] = iris_dataset['target']
df['target'] = df['target'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

display(df.head())
print(df.shape)
```

```python
# Построение диаграммы рассеяния

x = df['petal length (cm)']
y = df['petal width (cm)']

plt.scatter(x, y)

plt.xlabel('Длина лепестка')
plt.ylabel('Ширина лепестка')

plt.show()

# Либо так:

plt.scatter(x='petal length (cm)', y='petal width (cm)', data=df)

plt.xlabel('Длина лепестка')
plt.ylabel('Ширина лепестка')

plt.show()
```

```python
# Визуальное разделение точек на графике с использованием цикла

x = 'petal length (cm)'
y = 'petal width (cm)'

groups = df.groupby('target')

for name, group in groups:
    plt.scatter(group[x], group[y], label=name)

plt.legend()
plt.xlabel('Длина лепестка')
plt.ylabel('Ширина лепестка')

plt.show()
```

```python
# Визуальное разделение точек на графике без использования цикла

x1 = df[df['target'] == 'setosa']['petal length (cm)']
y1 = df[df['target'] == 'setosa']['petal width (cm)']

x2 = df[df['target'] == 'versicolor']['petal length (cm)']
y2 = df[df['target'] == 'versicolor']['petal width (cm)']

x3 = df[df['target'] == 'virginica']['petal length (cm)']
y3 = df[df['target'] == 'virginica']['petal width (cm)']

fig, ax = plt.subplots()

ax.scatter(x1, y1, label='setosa')
ax.scatter(x2, y2, label='versicolor')
ax.scatter(x3, y3, label='virginica')

plt.legend()
plt.xlabel('Длина лепестка')
plt.ylabel('Ширина лепестка')

plt.show()
```

```python
# Изменение отображения точек на графике

# edgecolor - цвет границ точек
# linewidth - ширина границ точек
# alpha - прозрачность точек

x = 'petal length (cm)'
y = 'petal width (cm)'

groups = df.groupby('target')

for name, group in groups:
    plt.scatter(group[x], group[y], label=name, edgecolor='black',  linewidth=0.9, alpha=0.8)

plt.legend()
plt.xlabel('Длина лепестка')
plt.ylabel('Ширина лепестка')

plt.show()
```

```python
# Параметры c и cmap

# Параметр c принимает на вход список или массив чисел и далее, в зависимости от выбранной цветовой карты cmap, 
# из этой цветовой карты выбирается тот цвет, который соответствует числу параметра c.

# Функция plt.colorbar() - шкала цвета
```

```python
x = df['petal length (cm)']
y = df['petal width (cm)']
z = df['petal width (cm)'] / df['petal length (cm)']

plt.scatter(x, y, c=z, cmap='summer')
plt.xlabel('Длина лепестка')
plt.ylabel('Ширина лепестка')

cbar = plt.colorbar()
cbar.set_label('Отношение ширины лепестка к его длине')

plt.show()
```

```python
# В вашем коде три графика (scatter) отображаются в разных цветах автоматически. Это поведение по умолчанию в Matplotlib.
# Когда вы вызываете plt.scatter() или любую другую функцию для построения графика без указания цвета, 
# Matplotlib автоматически выбирает цвет для каждого нового графика из стандартной палитры цветов. 
# Эта палитра определяет последовательность цветов, которые будут использоваться для различения разных наборов данных на графике.
# Таким образом, каждый вызов plt.scatter() в вашем коде создает новый набор точек, 
# и Matplotlib автоматически назначает каждому из них следующий цвет из палитры, чтобы графики были легко различимы. 
# Вы также можете явно задать цвет, используя параметр color или c внутри функции scatter()

x1 = [10,15,16,11,17,16,14,13,12]
y1 = [ 1, 3, 4, 2, 1, 5, 4, 3, 5]
x2 = [20,25,23,24,26,21,22,23,22]
y2 = [ 8,10, 9,11,12,10,14,10,13]
x3 = [26,30,24,28,24,28,31,32,27]
y3 = [ 3, 4, 6, 8, 7, 6, 8, 9, 2]

plt.scatter(x1, y1)
plt.scatter(x2, y2)
plt.scatter(x3, y3)

plt.show()
```

```python
# В этом коде все графики отображаются одним цветом, потому что вы передаете список списков в plt.scatter(). 
# Вместо того чтобы создавать три отдельных набора точек, plt.scatter([x1, x2, x3], [y1, y2, y3]) интерпретирует эти списки как один набор данных и отображает их в одном цвете.

# Как работает этот код:
# Аргументы x и y: В plt.scatter([x1, x2, x3], [y1, y2, y3]) аргументы x и y передаются в виде списков списков. 
# Функция scatter ожидает, что каждый аргумент (x и y) будет одномерным списком координат. 
# Но когда вы передаете список списков, Matplotlib воспринимает это как одну большую коллекцию точек, а не три отдельных набора.
# Результат: В итоге Matplotlib строит все точки на графике как один набор данных, 
# поэтому все точки оказываются одного цвета, который выбирается из стандартной палитры.

x1 = [10,15,16,11,17,16,14,13,12]
y1 = [ 1, 3, 4, 2, 1, 5, 4, 3, 5]
x2 = [20,25,23,24,26,21,22,23,22]
y2 = [ 8,10, 9,11,12,10,14,10,13]
x3 = [26,30,24,28,24,28,31,32,27]
y3 = [ 3, 4, 6, 8, 7, 6, 8, 9, 2]

plt.scatter([x1, x2, x3], [y1, y2, y3])

plt.show()
```

# 3.5 pie - круговыя диаграмма

# 3.6 stackplot - стековый график

```python
# Суть стекового графика заключается в том, что графики отображаются друг над другом, 
# и каждый следующий является суммой предыдущего и текущего наборов данных.  
# Такие графики с накопительным эффектом рисуются с помощью функции plt.stackplot()
```

```python
df_values = [['Понедельник', 300, 210, 157],
             ['Вторник', 295, 230, 180],
             ['Среда', 303, 214, 168],
             ['Четверг', 283, 254, 204],
             ['Пятница', 260, 263, 225],
             ['Суббота', 190, 177, 162],
             ['Воскресенье', 184, 181, 173]]
df_columns = ['День недели', 'Цех №1', 'Цех №2', 'Цех №3']
df = pd.DataFrame(df_values, columns=df_columns)

display(df)
```

```python
y1 = df['Цех №1']
y2 = df['Цех №2']
y3 = df['Цех №3']
x = df['День недели']
labels = ['Цех №1', 'Цех №2', 'Цех №3']
colors = ['#05B893', '#F59FC6', '#D7F59F']

plt.figure(figsize=(8, 5))

plt.stackplot(x, y1, y2, y3, 
              labels=labels,
              colors=colors,
              hatch='/*')             # - узор штриховки

plt.ylim(0, 1000)
plt.grid(alpha=0.3)
plt.legend()
plt.rcParams['hatch.color'] = 'white' # - цвет штриховки

plt.show()
```
