```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

# 1 Глава

```python
values = {'x': [2, 4, 8, 12, 15, 20, 25], 'y': [0, 5, 4, 11, 12, 10, 9]}
df = pd.DataFrame(values)

display(df)
```

# plt.plot() - построение графика

```python
plt.plot(df['x'], df['y'], color='purple', linewidth=3.5)
plt.show()
```

# plt.grid() - сетка

```python
plt.plot(df['x'], df['y'])
plt.grid()
plt.show()
```

```python
plt.plot(df['x'], df['y'])
plt.grid(color='green', linestyle=':', linewidth=2, axis='y')
plt.show()
```

# plt.title() - заголовок графика

```python
plt.plot(df['x'], df['y'], color='green')
plt.grid(color='grey', linestyle='--', linewidth=0.9)

plt.title('График зависимости признаков х и у', fontsize=18, color='purple', alpha=0.8)

plt.show()
```

# plt.xlabel(), plt.ylabel() - названия осей

```python
plt.plot(df['x'], df['y'])
plt.grid(color='grey', linestyle='--', linewidth=0.5)
plt.title('График зависимости признаков х и у')

plt.xlabel(df.columns[0], labelpad=20, color='red', fontsize=18)
plt.ylabel(df.columns[1], labelpad=50, rotation=0)

plt.show()
```

# plt.xlim, plt.ylim - изменение размерности осей графика

```python
plt.plot(df['x'], df['y'])
plt.grid(color='grey', linestyle='--', linewidth=0.5)
plt.title('График зависимости признаков х и у', fontsize=14)
plt.xlabel(df.columns[0], fontsize=14, color='red')
plt.ylabel(df.columns[1], fontsize=14, color='red')

plt.xlim(5, 15)
plt.ylim(2, 14)

plt.show()
```

# plt.xticks(), plt.yticks() - принудительно установить свои значения на осях графика

```python
plt.plot(df['x'], df['y'])
plt.grid(color='grey', linestyle='--', linewidth=0.5)
plt.title('График зависимости признаков х и у', fontsize=14)
plt.xlabel(df.columns[0], fontsize=14, color='red')
plt.ylabel(df.columns[1], fontsize=14, color='red')

plt.xticks(range(0, 26))
plt.yticks([0, 1, 2, 3, 4, 5, 10], rotation=45)

plt.show()
```

# color, linewidth, linestyle - цвет, ширина и типы линий графика

```python
plt.plot(df['x'], df['y'],   color='lime',      linewidth=1, linestyle='-')
plt.plot(df['x'], df['y']+2, color='gold',      linewidth=2, linestyle='--')
plt.plot(df['x'], df['y']+4, color='royalblue', linewidth=3, linestyle='-.')
plt.plot(df['x'], df['y']+6, color='olive',     linewidth=4, linestyle=':')

plt.show()

# Можно использовать сокращенные наименования параметров:
plt.plot(df['x'], df['y'], c='r', lw=2, ls='--');
```

# marker, markersize, markeredgecolor, markeredgewidth, markerfacecolor - варианты отображения точек на графике

```python
# '.', '*', 's', 'x', 'X', 'P', 'p', 'D', '<', '>', '^', 'v'
```

```python
plt.plot(df['x'], df['y'], color='r', linewidth=3, marker='o', markeredgecolor='b',
         markeredgewidth=4, markersize=20)

plt.show()
```

```python
plt.plot(df['x'], df['y'], color='green', linewidth=3, marker='o', markerfacecolor='red', 
         markeredgecolor='blue', markeredgewidth=5, markersize=15)

plt.show()
```

```python
plt.plot(df['x'], df['y'], color='red', alpha=0.5, marker='*', markersize=15, linestyle='None')

plt.show()
```

# Строим несколько линий на одном графике

```python
values = {'x1': [2, 4, 8, 12, 15, 20, 25], 
          'x2': [1, 3, 7, 10, 13, 18, 22],
          'y1': [0, 5, 4, 11, 12, 10, 9],
          'y2': [3, 10, 15, 6, 5, 2, 1]}

df = pd.DataFrame(values)

display(df)
```

```python
plt.plot(df['x1'], df['y1'], df['x2'], df['y2'], linewidth=3, marker='o')

plt.show()
```

```python
plt.plot(df['x1'], df['y1'], df['x2'], df['y2'], df['x2']+2, df['y2']+2, linewidth=3, marker='o')

plt.show()
```

```python
plt.plot(df['x1'],   df['y1'],   linewidth=3, marker='o')

plt.show()

plt.plot(df['x2'],   df['y2'],   linewidth=3, marker='o')
plt.plot(df['x2']+2, df['y2']+2, linewidth=3, marker='o')

plt.show()
```

```python
x = [0, 1, 2, 3, 4, 5]
y1 = x
y2 = [_**2 for _ in x]
y3 = [_**3 for _ in x]

plt.plot(x, y1, y2, y3)

plt.grid(ls='--', alpha=0.5)

plt.show()
```

```python
x = [0, 1, 2, 3, 4, 5]
y1 = x
y2 = [_**2 for _ in x]
y3 = [_**3 for _ in x]

plt.plot(x, y1, x, y2, x, y3)

plt.grid(ls='--', alpha=0.5)

plt.show()
```

# plt.legend() - отображение легенды

```python
plt.plot(df['x1'], df['y1'], linewidth=3, marker='o', label='x1_y1')
plt.plot(df['x2'], df['y2'], linewidth=3, marker='o', label='x2_y2')

plt.legend()

plt.show()
```

```python
# Параметр loc (расположение леггенды) функции plt.legend(), который может принимать следующие значения:

# 'best'         или 0
# 'upper right'  или 1
# 'upper left'   или 2
# 'lower left'   или 3
# 'lower right'  или 4
# 'right'        или 5
# 'center left'  или 6
# 'center right' или 7
# 'lower center' или 8
# 'upper center' или 9
# 'center'       или 10
```

```python
plt.plot(df['x1'], df['y1'], df['x2'], df['y2'], linewidth=3, marker='o')

plt.legend(['x1_y1', 'x2_y2'], loc=2)

plt.show()
```

```python
plt.plot(df['x1'], df['y1'], df['x2'], df['y2'], linewidth=3, marker='o')

plt.legend(['x1_y1', 'x2_y2'], loc=[0.5, 0.5]) # - точки в процентах от осей (50%, 50%)

plt.show()
```

```python
plt.plot(df['x1'], df['y1'], df['x2'], df['y2'], linewidth=3, marker='o')

plt.legend(['x1_y1', 'x2_y2'], loc=(1.05, 0.45))

plt.show()
```

# plt.subplot() - графики сразу на нескольких координатных осях

```python
x = np.linspace(0, 10, 5)
y1 = x**2
y2 = x**3
y3 = x**0.5
y4 = np.sin(x**2)
```

```python
plt.subplot(2, 2, 1) # - графики в 2 ряда, в графики должны отобразиться 2 столбцами и позиция конкретного подграфика (1, 2, 3, 4)
plt.plot(x, y1, 'r')
plt.title('y1')

plt.subplot(2, 2, 2)
plt.plot(x, y2, 'g')
plt.title('y2')

plt.subplot(2, 2, 3)
plt.plot(x, y3, 'y')
plt.title('y3')

plt.subplot(2, 2, 4)
plt.plot(x, y4, 'c')
plt.title('y4')

plt.show()
```

```python
# plt.tight_layout() - для устранения проблемы наложения данных друг на друга применяют функцию
# plt.suptitle() - общий заголовок 

plt.subplot(2, 2, 1)
plt.plot(x, y1, 'r')
plt.grid(':')
plt.title('y1')

plt.subplot(2, 2, 2)
plt.plot(x, y2, 'g')
plt.title('y2')

plt.subplot(2, 2, 3)
plt.plot(x, y3, 'y')
plt.grid()
plt.title('y3')

plt.subplot(2, 2, 4)
plt.plot(x, y4, 'c')
plt.grid('-.')
plt.title('y4')

plt.suptitle('Графики')

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(12, 7))

plt.subplot(4, 4, 1)
plt.plot(x, y1, 'r')
plt.subplot(4, 4, 2)
plt.plot(x, y2, 'b')
plt.subplot(4, 4, 3)
plt.plot(x, y3, 'g')
plt.subplot(4, 4, 4)
plt.plot(x, y4, 'y')

plt.subplot(4, 3, 4)
plt.plot(x, y1, 'm')
plt.subplot(4, 3, 5)
plt.plot(x, y2, 'c')
plt.subplot(4, 3, 6)
plt.plot(x, y3, 'orange')

plt.subplot(4, 2, 5)
plt.plot(x, y4, 'grey')
plt.subplot(4, 2, 6)
plt.plot(x, y1, 'brown')

plt.subplot(4, 1, 4)
plt.plot(x, y2)


plt.tight_layout()

plt.show()
```

# figure(), axes(). Наносим координатные оси на график

```python
# Figure - это холст, на котором мы будем размещать координатные оси (объекты Axes)
# figsize - задаёт размер всей фигуры,
# plt.axes([x0, y0, ширина, высота]) - управляет размерами и положением осей внутри фигуры, 
# где ширина и высота - это доли от размеров фигуры.
```

```python
plt.figure(figsize=(8,6), dpi=200) # размер и разрешение

plt.show()
```

```python
fig = plt.figure(figsize=(7,5), facecolor='olive') # - заливка области объекта Figure

ax = plt.axes([0, 0, 1, 1]) # 0, 0 - начало координат; 1, 1 - конечные точки на графике

plt.show()
```

```python
fig = plt.figure(figsize=(7,5), facecolor='olive')

ax1 = plt.axes([0, 0, 0.5, 0.5])
ax2 = plt.axes([0.5, 0.5, 0.5, 0.5])
plt.show()
```

```python
fig = plt.figure(figsize=(7,5), facecolor='olive')

ax1 = plt.axes([0, 0, 0.5, 0.5])
ax2 = plt.axes([0.5, 0.5, 0.8, 0.8])

plt.show()
```

```python
fig = plt.figure(figsize=(7,5), facecolor='olive')

ax = fig.add_axes([0, 0, 1, 1]) # == plt.axes()

plt.show()
```

```python
fig = plt.figure(figsize=(5, 5), facecolor='lightgrey')

ax1 = plt.axes([0, 0, 0.4, 0.4])        # - левый нижний
ax2 = plt.axes([0.5, 0.5, 0.4, 0.4])    # - правый верхний
ax3 = plt.axes([0, 0.5, 0.4, 0.4])      # - левый верхний

plt.show()
```
