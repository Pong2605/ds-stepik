```python
concentrations = pd.read_csv('data/algae.csv')
concentrations.head()
```

```python
mean_concentrations = concentrations.groupby(['genus']).aggregate({'sucrose':'mean', 'alanin':'mean', 'citrate':'mean', 'glucose':'mean', 'oleic_acid':'mean'})
mean_concentrations.head()
```

```python
mean_concentrations = concentrations.groupby('genus').mean(numeric_only=True)
mean_concentrations.head()
```

```python
mean_concentrations = concentrations.groupby('genus')[['sucrose','alanin','citrate','glucose','oleic_acid']].mean()
mean_concentrations.head()
```

```python
concentrations.query("genus == 'Fucus'").alanin.describe().loc[['min', 'mean', 'max']].round(2)
```

```python
concentrations.query('genus=="Fucus"').agg({'alanin': ['min', 'mean', 'max']}).round(2)
```

```python
concentrations.groupby('group').agg({'sucrose': ['min', 'max'], 'species': 'nunique', 'citrate' : 'var'})
```

```python
concentrations.groupby('group').aggregate({'sucrose':'max'})['sucrose'] - concentrations.groupby('group').aggregate({'sucrose':'min'})['sucrose']
```

```python
def razmah(lab):
    mn = np.max(lab) - np.min(lab)
    return mn
```

```python
concentrations.groupby('group').agg({'sucrose' : razmah})
```
