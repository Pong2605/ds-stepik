```python
# Напишите функцию calculate_age(birth_date), 
# которая принимает на вход дату рождения пользователя в формате (год, месяц, день) 
# и возвращает его возраст в полных годах.

from datetime import datetime

def calculate_age(birth_date):
    age = datetime.now() - datetime(*birth_date)
    return age.days // 365

def test_calculate_age():
    # Тестовые входные данные и ожидаемые результаты
    test_cases = [
        ((2000, 1, 1), 24),  # Тест на возраст 24 года
        ((1985, 12, 31), 38),  # Тест на возраст 36 лет
        ((2023, 6, 30), 0),  # Тест на возраст 1 год
        ((2010, 2, 28), 13),  # Тест на возраст 14 лет
        ((2012, 2, 29), 11)   # Тест на возраст 12 лет (високосный год)
    ]

    # Проверяем каждый тестовый случай
    for birth_date, expected_age in test_cases:
        # Вычисляем возраст
        age = calculate_age(birth_date)
        expected_age += (datetime.now().year - 2024)
        # Проверяем, что результат совпадает с ожидаемым
        assert age == expected_age, f"Ошибка: Для даты {birth_date} ожидается возраст {expected_age}, но получено {age}"

# Запуск тестов
test_calculate_age()
print("Все тесты пройдены успешно!")
```

```python
# В модуле должна быть функция solve_quadratic_equation(a, b, c), 
# принимающая три аргумента - коэффициенты квадратного уравнения
# и возвращающая корни уравнения.
# В случае отсутствия действительных корней, верните сообщение:  Корни не существуют

def solve_quadratic_equation(a, b, c):
    x1 = (-b + (b**2 - 4*a*c)**0.5) / (2*a)
    x2 = (-b - (b**2 - 4*a*c)**0.5) / (2*a)
    if isinstance(x1, complex) or isinstance(x2, complex):
        return 'Корни не существуют'
    return x1, x2

# from quadratic_solver import solve_quadratic_equation

# Тестирование
coefficients1 = (1, -3, 2)
coefficients2 = (1, 2, 1)
coefficients3 = (1, 1, 1)

result1 = solve_quadratic_equation(*coefficients1)
result2 = solve_quadratic_equation(*coefficients2)
result3 = solve_quadratic_equation(*coefficients3)

print(f"Уравнение {coefficients1}: Корни = {result1}")
print(f"Уравнение {coefficients2}: Корни = {result2}")
print(f"Уравнение {coefficients3}: Корни = {result3}")
```

```python
def covid19_stat(covid_cases_last_7_days):
    cases = covid_cases_last_7_days.values()
    report = f'''Описательная статистика за последние 7 дней:
Всего случаев: {sum(cases)}
Среднее количество случаев в день: {sum(cases) / 7}
Максимальное количество случаев в день: {max(cases)}
Минимальное количество случаев в день: {min(cases)}'''
    print(report)

covid_cases_last_7_days = {
    '2024-02-08': 500,
    '2024-02-09': 600,
    '2024-02-10': 700,
    '2024-02-11': 800,
    '2024-02-12': 900,
    '2024-02-13': 1000,
    '2024-02-14': 1100
}

covid19_stat(covid_cases_last_7_days)
```

```python
# Воспользовавшись функцией walk, можно получить доступ к названиям и путям всех подпапок и файлов, 
# относящихся к заданному каталогу. Применив один внешний цикл for, 
# а также два вложенных, несложно получить информацию об объектах в каталоге folder через специальные списки directories и files. 
# Сведения выдаются с помощью многократного обращения к print.

for root, directories, files in os.walk(r"./"):
    print(root)
    for directory in directories:
        print(directory)
    for file in files:
        print(file)
```
