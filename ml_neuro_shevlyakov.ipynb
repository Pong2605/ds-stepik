```python
import sympy as sp
import numpy as np

class GradientDescent:
    def __init__(self, f, a, h, n_iter, mode='simple', T=None, eps=1e-8, b1=0.9, b2=0.9):
        self.f = f
        self.symbols = sorted(sp.sympify(f).free_symbols, key=str)  # x1, x2, ...
        self.a = np.array(a, dtype=float)
        self.h = h
        self.T = T
        self.n_iter = n_iter
        self.mode = mode
        self.eps = eps
        self.b1 = b1
        self.b2 = b2
        
        # Символьное выражение
        self.f_expr = sp.sympify(self.f)
        
        # Для AdaGrad: сумма квадратов градиентов
        if self.mode == 'adagrad':
            self.G = np.zeros(len(self.symbols))

         # Для Adam: инициализация моментов
        if self.mode == 'adam':
            self.m = np.zeros(len(self.symbols))  # первый момент
            self.v = np.zeros(len(self.symbols))  # второй момент
            self.t = 0  # счётчик шагов   

    def diff(self):
        """Вычисляет градиент функции f по всем переменным."""
        return [sp.diff(self.f_expr, var) for var in self.symbols]
    
    def h_simple(self, h, n):
        """Линейное уменьшение шага."""
        return h * (1 - n / self.T) if self.T else h

    def h_simple2(self, h, n):
        """Экспоненциальное уменьшение шага."""
        return h * float(sp.exp(-n / self.T))

    def fit(self):
        a = self.a.copy()
        grad = self.diff()  # Список символьных производных
        
        for i in range(self.n_iter):
            # Вычисляем градиент в точке a
            grad_vals = np.array([
                g.subs(dict(zip(self.symbols, a))).evalf()
                for g in grad
            ], dtype=float)
    
            # Определяем текущий шаг h
            if self.mode == 'adagrad':
                # AdaGrad: адаптивный шаг по компонентам
                self.G += grad_vals ** 2
                current_h = self.h / np.sqrt(self.G + self.eps)

            elif self.mode == 'adam':
                # Шаг Adam
                self.t += 1
                # Обновление первого момента
                self.m = self.b1 * self.m + (1 - self.b1) * grad_vals

                # Обновление второго момента
                self.v = self.b2 * self.v + (1 - self.b2) * (grad_vals ** 2)

                # Коррекция смещения
                m_hat = self.m / (1 - self.b1 ** self.t)
                v_hat = self.v / (1 - self.b2 ** self.t)

                # Обновление параметров
                current_h = self.h * m_hat / (np.sqrt(v_hat) + self.eps)
                a = a - current_h
            else:
                # Другие режимы: скалярный шаг
                h_func = getattr(self, 'h_' + self.mode)
                current_h = h_func(self.h, i)

            # Обновляем параметры
            a = a - current_h * grad_vals

            # Вывод (округляем для читаемости)
            a_rounded = np.round(a, 6)
            if isinstance(current_h, np.ndarray):
                h_rounded = np.round(current_h, 6)
            else:
                h_rounded = round(current_h, 6)
            print(f'Iteration {i}: a = {a_rounded}, h = {h_rounded}')
            
        return a


f = 'x1**2'
a = [1.0]
h = 1
T = 4
n_iter = 5

optimizer = GradientDescent(f, a, h, n_iter, mode='simple', T=T)
result = optimizer.fit()
print(f'Final result: {result}')
```

```python
import sympy as sp
import numpy as np

class GradientDescent:
    def __init__(self, f, a, h, n_iter, mode='simple', T=None, eps=1e-8, b1=0.9, b2=0.9):
        self.f = f
        self.symbols = sorted(sp.sympify(f).free_symbols, key=str)  # x1, x2, ...
        self.a = np.array(a, dtype=float)
        self.h = h
        self.T = T
        self.n_iter = n_iter
        self.mode = mode
        self.eps = eps
        self.b1 = b1
        self.b2 = b2
        
        # Символьное выражение
        self.f_expr = sp.sympify(self.f)
        
        # Для Adam: инициализация моментов
        if self.mode == 'adam':
            self.m = np.zeros(len(self.symbols))  # первый момент
            self.v = np.zeros(len(self.symbols))  # второй момент
            self.t = 0  # счётчик шагов

    def diff(self):
        """Вычисляет градиент функции f по всем переменным."""
        return [sp.diff(self.f_expr, var) for var in self.symbols]

    def h_simple(self, h, n):
        """Линейное уменьшение шага."""
        return h * (1 - n / self.T) if self.T else h

    def h_simple2(self, h, n):
        """Экспоненциальное уменьшение шага."""
        return h * float(sp.exp(-n / self.T))

    def fit(self):
        a = self.a.copy()
        grad = self.diff()  # Список символьных производных

        for i in range(self.n_iter):
            # Вычисляем градиент в точке a
            grad_vals = np.array([
                g.subs(dict(zip(self.symbols, a))).evalf()
                for g in grad
            ], dtype=float)

            if self.mode == 'adam':
                # Шаг Adam
                self.t += 1

                # Обновление первого момента
                self.m = self.b1 * self.m + (1 - self.b1) * grad_vals

                # Обновление второго момента
                self.v = self.b2 * self.v + (1 - self.b2) * (grad_vals ** 2)

                # Коррекция смещения
                m_hat = self.m / (1 - self.b1 ** self.t)
                v_hat = self.v / (1 - self.b2 ** self.t)

                # Обновление параметров
                step = self.h * m_hat / (np.sqrt(v_hat) + self.eps)
                a = a - step

                current_h_info = f"m_hat={np.round(m_hat,6)}, v_hat={np.round(v_hat,6)}"

            else:
                # Другие режимы (simple, simple2 и т.п.)
                h_func = getattr(self, 'h_' + self.mode)
                current_h = h_func(self.h, i)
                a = a - current_h * grad_vals
                current_h_info = round(current_h, 6)

            # Вывод результата
            a_rounded = np.round(a, 6)
            print(f'Iteration {i}: a = {a_rounded}, h_info = {current_h_info}')

        return a



# Пример: минимизация f(x1,x2) = x1^2 + 10*x2^2
f = 'x1**2 + 10 * x2**2'
a0 = [10.0, 1.0]
h = 0.1
n_iter = 2

# Запускаем Adam
optimizer = GradientDescent(f, a0, h, n_iter, mode='adam', eps=0, b1=0.9, b2=0.9)
result = optimizer.fit()

print(f'\nFinal result: {np.round(result, 6)}')
print(f'Первая координата a2: {result[0]:.4f}')
```
